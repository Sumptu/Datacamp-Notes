# Matplotlib

```python
import matplotlib.pyplot as plt

year = [1950, 1970, 1990, 2010]
pop = [2.519, 3.692, 5.263, 6.972]
```

```python
plt.plot(year, pop)
plt.show()
```

```python
plt.scatter(year, pop)
plt.show()
```
