**Tugas:** Buatlah sebuah fungsi bernama `proses_log_suhu`. Fungsi ini menerima satu argumen, yaitu sebuah *list of strings* yang berisi data log suhu yang tidak rapi.

Fungsi ini harus melakukan hal-hal berikut di dalamnya:
1.  Menggunakan *selective import* untuk mengimpor fungsi `ceil` dari *package* `math`.
2.  Untuk setiap *string* di dalam *list* input:
    * Hilangkan spasi kosong di awal dan akhir *string* menggunakan *string method*.
    * Pisahkan *string* menjadi dua bagian berdasarkan karakter `;`.
    * Ambil bagian kedua (suhu), hilangkan teks `" C"`, ubah menjadi `float`.
    * Bulatkan angka suhu tersebut **ke atas** ke bilangan bulat terdekat menggunakan fungsi `ceil` yang sudah diimpor.
3.  Fungsi harus mengembalikan sebuah *list* baru yang hanya berisi angka-angka suhu yang sudah dibulatkan sebagai *integer*.

**Contoh Pemanggilan dan Output:**
```python
# List input
log_data_h6 = [' ID-001; 24.3 C ', '  ID-002; 25.9 C', 'ID-003; 25.0 C  ']

# Memanggil fungsi
suhu_terproses = proses_log_suhu(log_data_h6)
print(suhu_terproses)

# Output yang diharapkan
[25, 26, 25]

````python
from math import ceil

def proses_log_suhu(data_kotor):
    data_bersih = [item.strip() for item in data_kotor]
    data_bersih = [item.split(';') for item in data_bersih]
    data_bersih = [item[1] for item in data_bersih]
    data_bersih = [item.replace('C', '') for item in data_bersih]
    data_bersih = [float(item) for item in data_bersih]
    data_bersih = [ceil(item) for item in data_bersih]
    return data_bersih

log_data_h6 = [' ID-001; 24.3 C ', '  ID-002; 25.9 C', 'ID-003; 25.0 C  ']

suhu_terproses = proses_log_suhu(log_data_h6)
print(suhu_terproses)
````

1.  Fungsi ini menerima satu argumen: sebuah list berisi angka (bisa *integer* atau *float*).
2.  Di dalam fungsi, **impor *package* `numpy` dengan alias `np`**.
3.  Ubah list input tersebut menjadi sebuah NumPy *array*.
4.  Gunakan fungsi dari NumPy untuk menghitung **mean** (rata-rata) dan **standard deviation** (deviasi standar) dari *array* tersebut.
5.  Fungsi harus **mengembalikan kedua nilai** tersebut secara berurutan: (mean, standard deviation).

**Contoh Pemanggilan dan Output:**
```python
# List nilai ujian
nilai_ujian_h6 = [85, 90, 78, 92, 88, 76, 89, 95, 80]

````python
def analisis_statistik_dasar(angka):
    import numpy as np
    np_angka = np.array(angka)
    mean = np.mean(np_angka)
    std_dev = np.std(np_angka)
    return mean, std_dev

nilai_ujian_h6 = [85, 90, 78, 92, 88, 76, 89, 95, 80]

mean_ujian, std_dev_ujian = analisis_statistik_dasar(nilai_ujian_h6)

print(f"Mean: {mean_ujian:.2f}")
print(f"Standar Deviation: {std_dev_ujian:.2f}")
````

Fungsi ini harus melakukan hal berikut:

2.  Memisahkan setiap nama yang sudah diformat menjadi nama depan dan nama belakang. Asumsikan setiap nama hanya terdiri dari dua kata.
3.  Fungsi harus mengembalikan sebuah **dictionary** dengan dua *key*: `'nama_depan'` dan `'nama_belakang'`.
    * *Value* untuk *key* `'nama_depan'` adalah sebuah list yang berisi semua nama depan.
    * *Value* untuk *key* `'nama_belakang'` adalah sebuah list yang berisi semua nama belakang.

**Contoh Pemanggilan dan Output:**
```python
# List nama dengan format acak
daftar_nama_h6 = ['budi santoso', '  ANI WIDJAJA', 'susilo bambang  ', 'MEGAWATI putri']

{'nama_depan': ['Budi', 'Ani', 'Susilo', 'Megawati'], 'nama_belakang': ['Santoso', 'Widjaja', 'Bambang', 'Putri']}

1.  `daftar_email` (sebuah *list* berisi alamat email).
2.  `provider_target` (sebuah *string*, yang merupakan nama domain yang ingin dicari, contoh: `"gmail.com"`). Argumen ini harus memiliki **nilai default** `"gmail.com"`.

Fungsi ini harus melakukan hal berikut:
1.  Iterasi melalui setiap email dalam `daftar_email`.
2.  Untuk setiap email, periksa apakah email tersebut berasal dari `provider_target`. Lakukan ini dengan mengambil bagian *string* setelah simbol `@`.
3.  Fungsi harus mengembalikan sebuah *list* baru yang hanya berisi email-email yang cocok dengan `provider_target`.
4.  Semua email dalam *list* yang dikembalikan harus dalam format **huruf kecil (lowercase)**.

**Contoh Pemanggilan dan Output:**
```python
# List email dengan format beragam
list_email_h6 = ["Budi.Susanto@Gmail.com", "ani@yahoo.com", "Cecep@GMAIL.COM", "dodi@outlook.com", "eko@yahoo.co.id"]

# 1. Memanggil fungsi TANPA memberikan argumen kedua (menggunakan default)

**Tugas:** Buatlah sebuah fungsi bernama `hitung_sisa_hari`. Fungsi ini menerima satu argumen: sebuah *list* yang berisi *string* tanggal dengan format `"YYYY-MM-DD"`.

Fungsi ini harus melakukan hal berikut:
1.  Menggunakan *selective import* untuk mengimpor *class* `datetime` dari *package* `datetime`.
2.  Di dalam fungsi, dapatkan tanggal dan waktu **saat ini** menggunakan `datetime.now()`.
3.  Untuk setiap *string* tanggal di dalam *list* input:
    * Ubah (*parse*) *string* tersebut menjadi sebuah `datetime` *object*. *Hint: Gunakan `datetime.strptime(string_tanggal, format)`.*
    * Hitung selisih hari antara tanggal dari *list* dengan tanggal hari ini. Hasil dari pengurangan dua `datetime` *object* adalah sebuah `timedelta` *object*. Anda bisa mendapatkan harinya dengan mengakses atribut `.days`.
4.  Fungsi harus mengembalikan sebuah *list* baru yang berisi sisa hari dalam bentuk *integer*. Nilainya bisa negatif jika tanggalnya sudah berlalu.

**Contoh Pemanggilan dan Output:**
```python
# Impor dulu untuk menjalankan contoh
from datetime import datetime

# Daftar tanggal acara
list_acara_h6 = ["2025-12-25", "2025-07-20", "2025-01-01"]

# Memanggil fungsi
sisa_hari = hitung_sisa_hari(list_acara_h6)
print(sisa_hari)

# Output yang diharapkan akan bervariasi tergantung kapan kode dijalankan.
# Jika kode dijalankan pada 9 Juli 2025, outputnya akan seperti ini:
# [169, 11, -189]

````python
from datetime import datetime

def hitung_sisa_hari(tanggal):
    tanggal_sekarang = datetime.now()
    hasil_selisih = []
    for i in tanggal:
        tanggal_acara = datetime.strptime(i, "%Y-%m-%d")
        selisih = tanggal_acara - tanggal_sekarang
        hasil_selisih.append(selisih.days)
    return hasil_selisih

list_acara_h6 = ["2025-12-25", "2025-07-20", "2025-01-01"]

sisa_hari = hitung_sisa_hari(list_acara_h6)
print(sisa_hari)
````

4.  Fungsi harus mengembalikan sebuah **list baru** yang berisi *dictionary-dictionary* produk yang sudah bersih dan terfilter.

**Contoh Pemanggilan dan Output:**
```python
# Daftar produk dengan data kotor
data_produk_h6 = [
    {'nama': 'buku tulis', 'harga': ' Rp 5000 ', 'stok': 10},
    {'nama': 'PENSIL 2B', 'harga': 'Rp 2500', 'stok': 4},
    {'nama': '  penghapus  ', 'harga': 'Rp 1500', 'stok': 20},

````python
def proses_data_produk(list_produk, stok_minimum=5):
    lolos_filter = []

    for produk in list_produk:
        if produk["stok"] >= stok_minimum:
            nama_title = produk["nama"].strip().title()
            harga_int = int(produk["harga"].replace("Rp", "").strip())
            produk_baru = {
                "nama": nama_title,
                "harga": harga_int,
                "stok": produk["stok"]
            }
            lolos_filter.append(produk_baru)
    return lolos_filter
# Daftar produk dengan data kotor
data_produk_h6 = [
    {'nama': 'buku tulis', 'harga': ' Rp 5000 ', 'stok': 10},
    {'nama': 'PENSIL 2B', 'harga': 'Rp 2500', 'stok': 4},
    {'nama': '  penghapus  ', 'harga': 'Rp 1500', 'stok': 20},
    {'nama': 'penggaris', 'harga': '  Rp 3000  ', 'stok': 8}
]

# Memanggil fungsi
produk_siap_jual = proses_data_produk(data_produk_h6)
print(produk_siap_jual)
````

### Soal 7 dari 10 (Versi Sulit)

**Topik Gabungan:** Error Handling (try-except), Full Package Import, Data Conversion

**Konteks:** Dalam dunia nyata, data seringkali "rusak". Sebuah kolom yang seharusnya berisi angka mungkin saja berisi teks. Kode kita harus bisa menangani ini tanpa berhenti total (*crash*).

**Tugas:** Buatlah sebuah fungsi bernama `jumlahkan_angka_valid`. Fungsi ini menerima satu argumen: sebuah *list* yang berisi *string* yang campur aduk (sebagian besar adalah angka, tapi beberapa mungkin teks acak).

Fungsi ini harus melakukan hal berikut:
1.  Impor seluruh *package* `math` (`import math`).
2.  Inisialisasi sebuah variabel untuk total (misal, `total_akhir = 0`).
3.  Lakukan iterasi pada setiap item di dalam *list* input.
4.  Di dalam *loop*, gunakan blok **`try...except`**.
    * Di dalam blok `try`: coba ubah item menjadi `float`. Jika berhasil, bulatkan angka tersebut **ke bawah** ke bilangan bulat terdekat menggunakan `math.floor()` dan tambahkan ke `total_akhir`.
    * Blok `except` harus menangkap `ValueError` (*error* yang terjadi jika konversi gagal). Jika *error* ini terjadi, cukup abaikan item tersebut dan lanjutkan ke item berikutnya (Anda bisa menggunakan `pass`).
5.  Setelah *loop* selesai, kembalikan nilai `total_akhir`.

**Contoh Pemanggilan dan Output:**
```python
# List data dengan angka valid dan data rusak
data_mentah_h6 = ["10.5", "20.1", "ini bukan angka", "30.9", "7", "error data"]

# Memanggil fungsi
total = jumlahkan_angka_valid(data_mentah_h6)
print(f"Total angka yang valid: {total}")

# Output yang diharapkan
Total angka yang valid: 67
# Penjelasan: floor(10.5) + floor(20.1) + floor(30.9) + floor(7) = 10 + 20 + 30 + 7 = 67

````python
import math

def jumlahkan_angka_valid(list_input):
    total_akhir = 0
    for i in list_input:
        try:
            float_i = float(i)
            total_akhir += math.floor(float_i)
        except ValueError:
            pass
    return total_akhir

# List data dengan angka valid dan data rusak
data_mentah_h6 = ["10.5", "20.1", "ini bukan angka", "30.9", "7", "error data"]

# Memanggil fungsi
total = jumlahkan_angka_valid(data_mentah_h6)
print(f"Total angka yang valid: {total}")
````

**Konteks:** Kadang-kadang, untuk kenyamanan, *programmer* mengimpor semua fungsi dari sebuah *package* menggunakan `*` (dikenal sebagai *star import*). Ini memungkinkan kita memanggil fungsi tanpa menyebut nama *package*-nya (misal, `floor()` bukan `math.floor()`). Meskipun nyaman, praktik ini seringkali **tidak disarankan**. **Tugas:** Buatlah sebuah fungsi bernama `hitung_luas_area`.
1.  Di bagian atas kode Anda (di luar fungsi), gunakan `from math import *` untuk mengimpor semua dari *package* `math`.
2.  Fungsi `hitung_luas_area` menerima tiga argumen:
    * `bentuk` (*string*: `"lingkaran"`, `"persegi"`, atau `"segitiga"`).
    * `nilai1` (*integer* atau *float*).
    * `nilai2` (*integer* atau *float*, dengan **nilai default 0**).
3.  Di dalam fungsi, gunakan `if/elif/else` untuk menghitung luas berdasarkan `bentuk`:
    * Jika `"lingkaran"`, `nilai1` adalah jari-jari. Luasnya adalah `pi * pow(nilai1, 2)`.
    * Jika `"persegi"`, `nilai1` adalah sisi. Luasnya adalah `pow(nilai1, 2)`.
    * Jika `"segitiga"`, `nilai1` adalah alas dan `nilai2` adalah tinggi. Luasnya adalah `0.5 * nilai1 * nilai2`.
4.  Fungsi harus mengembalikan hasil perhitungan luas.

**Contoh Pemanggilan dan Output:**
```python
# Memanggil fungsi untuk berbagai bentuk
luas_lingkaran = hitung_luas_area("lingkaran", 7)
luas_persegi = hitung_luas_area("persegi", 10)
luas_segitiga = hitung_luas_area("segitiga", 10, 5)

print(f"Luas Lingkaran: {luas_lingkaran:.2f}")
print(f"Luas Persegi: {luas_persegi}")
print(f"Luas Segitiga: {luas_segitiga}")

# Output yang diharapkan
Luas Lingkaran: 153.94
Luas Persegi: 100.0
Luas Segitiga: 25.0

````python
from math import *

def hitung_luas_area(bentuk, nilai1, nilai2=0):
    if bentuk == "lingkaran":
        return pi * pow(nilai1, 2)
    elif bentuk == "persegi":
        return pow(nilai1, 2)
    elif bentuk == "segitiga":
        return 0.5 * nilai1 * nilai2

    
# Memanggil fungsi untuk berbagai bentuk
luas_lingkaran = hitung_luas_area("lingkaran", 7)
luas_persegi = hitung_luas_area("persegi", 10)
luas_segitiga = hitung_luas_area("segitiga", 10, 5)

print(f"Luas Lingkaran: {luas_lingkaran:.2f}")
print(f"Luas Persegi: {luas_persegi}")
print(f"Luas Segitiga: {luas_segitiga}")
````

**Tugas:** Buatlah sebuah fungsi bernama `buat_data_kelulusan`. Fungsi ini menerima dua argumen:
1.  `jumlah_siswa` (*integer*).
2.  `batas_lulus` (*integer*, dengan nilai **default 60**).

Fungsi ini harus menghasilkan dan mengembalikan sebuah **list of dictionaries**, di mana setiap *dictionary* adalah satu siswa dengan data sebagai berikut:
1.  Impor *package* `random`.
2.  Lakukan *looping* sebanyak `jumlah_siswa` untuk membuat data bagi setiap siswa.
3.  Setiap siswa harus memiliki *dictionary* dengan *key* berikut:
    * `'id_siswa'`: Sebuah *string* dengan format `"STD-{nomor}"`, di mana nomornya adalah tiga digit dengan *zero-padding* (misal: `"STD-001"`, `"STD-002"`, ..., `"STD-010"`). *Hint: Gunakan f-string formatting seperti `{i:03d}`*.
    * `'nilai'`: Sebuah *integer* acak antara 0 dan 100 (inklusif), gunakan `random.randint(0, 100)`.
    * `'status'`: Sebuah *string*, `"Lulus"` jika `'nilai'` lebih besar atau sama dengan `batas_lulus`, dan `"Gagal"` jika sebaliknya.
4.  Kembalikan *list* yang berisi semua *dictionary* siswa tersebut.

**Contoh Pemanggilan dan Output:**
```python
# Membuat data untuk 5 siswa
data_siswa = buat_data_kelulusan(5)

# Mencetak hasil (output akan berbeda setiap kali karena ada unsur acak)
# Kita gunakan 'json' untuk mencetak dictionary dengan rapi
import json
print(json.dumps(data_siswa, indent=4))

# Contoh kemungkinan output:
# [
#     {
#         "id_siswa": "STD-001",
#         "nilai": 85,
#         "status": "Lulus"
#     },
#     {
#         "id_siswa": "STD-002",
#         "nilai": 42,
#         "status": "Gagal"
#     },
#     ... dan seterusnya untuk 5 siswa
# ]

````python
import json
import random

def buat_data_kelulusan(jumlah_siswa, batas_lulus=60):
    list_of_dict = []
    for data in range(jumlah_siswa):
        id_siswa = f"STD-{data + 1:03d}"
        nilai = random.randint(0, 100)
        status = "Lulus" if nilai >= batas_lulus else "Gagal"
        data_siswa = {
            "id_siswa": id_siswa,
            "nilai": nilai,
            "status": status
        }
        list_of_dict.append(data_siswa)
    return list_of_dict


# Membuat data untuk 5 siswa
data_siswa = buat_data_kelulusan(5)

# Mencetak hasil (output akan berbeda setiap kali karena ada unsur acak)
print(json.dumps(data_siswa, indent=4))
````

**Konteks:** Anda menerima data penjualan mentah dalam format *string* tunggal yang dipisahkan koma. Tugas Anda adalah membersihkan, memvalidasi, dan menganalisis data ini, lalu menyajikannya dalam format yang rapi.

**Tugas:** Buatlah sebuah fungsi bernama `analisis_penjualan_final`. Fungsi ini menerima dua argumen:
1.  `data_raw` (sebuah *list* berisi *string-string* penjualan).
2.  `diskon_persen` (sebuah *float* antara 0 dan 100, untuk menghitung harga setelah diskon).

Setiap *string* dalam `data_raw` memiliki format: `" IDProduk, Nama Produk , Jumlah, Harga Satuan "`. Contoh: `"P001,   Buku Tulis  , 10, 2500"`.

Fungsi ini harus melakukan hal berikut:
1.  Impor `numpy` dengan alias `np`.
2.  Siapkan sebuah *list* kosong untuk menampung hasil yang sudah bersih.
3.  Lakukan iterasi untuk setiap `string` di `data_raw`:
    a. Gunakan blok `try...except` untuk menangani baris data yang mungkin rusak (misalnya, jumlah bagian tidak pas, atau harga/jumlah bukan angka). Jika ada `Exception` apa pun, lewati baris tersebut.
    b. Di dalam `try`, pisahkan *string* berdasarkan koma (`,`). Pastikan ada **tepat 4 bagian**. Jika tidak, lewati baris ini.
    c. Bersihkan setiap bagian:
        * ID Produk: Hapus spasi.
        * Nama Produk: Hapus spasi dan ubah ke *Title Case*.
        * Jumlah: Hapus spasi, ubah ke *integer*.
        * Harga Satuan: Hapus spasi, ubah ke *integer*.
    d. Hitung dua nilai baru:
        * `total_harga`: `jumlah * harga_satuan`.
        * `harga_setelah_diskon`: `total_harga * (1 - (diskon_persen / 100))`.
    e. Buat sebuah *dictionary* untuk produk yang berhasil diproses berisi: `id`, `nama`, `jumlah`, `total_harga`, `harga_final`.
    f. Tambahkan *dictionary* ini ke *list* hasil.
4.  Setelah memproses semua data, jika *list* hasil tidak kosong, gunakan `numpy` untuk menghitung **rata-rata** dari `harga_final` semua produk.
5.  Fungsi harus mengembalikan sebuah **tuple** yang berisi dua hal:
    * *List* berisi *dictionary* produk yang bersih.
    * Rata-rata harga final (atau 0 jika tidak ada data yang valid).

**Contoh Pemanggilan dan Output:**

# Rata-rata Harga Final: 21000.00

````python
import json
import numpy as np

def analisis_penjualan_final(data_raw, diskon_persen):
    data_bersih = []
    for raw in data_raw:
        try:
            raw_split = raw.split(',')
            id_produk = raw_split[0].strip()
            nama_produk = raw_split[1].strip().title()
            jumlah = int(raw_split[2].strip())
            harga = int(raw_split[3].strip())

            total_harga = jumlah * harga
            harga_setelah_diskon = total_harga * (1 - diskon_persen/100)

            dicti = {
                "id": id_produk,
                "nama": nama_produk,
                "jumlah": jumlah,
                "total_harga": total_harga,
                "harga_final": harga_setelah_diskon,
            }
            data_bersih.append(dicti)
        except (ValueError, IndexError):
            pass
    rata_rata_harga_final = 0.0
    if data_bersih:
        rata_rata_harga_final = [item["harga_final"] for item in data_bersih]
        rata_rata_harga_final = np.mean(rata_rata_harga_final)

    return data_bersih, rata_rata_harga_final

# Data penjualan mentah dengan beberapa baris rusak
data_penjualan_h6 = [
    "P001,   Buku Tulis  , 10, 2500",
    "P002, Pensil 2B, 20, 1500", # Kurang spasi, harusnya tetap bisa
    "P003, PENGHAPUS, 15, 1000",
    "P004, Penggaris, ini bukan angka, 3000", # Data jumlah rusak
    "P005, Kertas A4, 5" # Data tidak lengkap
]

# Memanggil fungsi
produk_terjual, avg_harga_final = analisis_penjualan_final(data_penjualan_h6, 10.0) # diskon 10%

# Mencetak hasil
print("Data Terproses:")
print(json.dumps(produk_terjual, indent=2))
print(f"\nRata-rata Harga Final: {avg_harga_final:.2f}")
````

````python
import json

data_penjualan_h6 = [
    "P001,   Buku Tulis  , 10, 2500",
    "P002, Pensil 2B, 20, 1500", # Kurang spasi, harusnya tetap bisa
    "P003, PENGHAPUS, 15, 1000",
    "P004, Penggaris, ini bukan angka, 3000", # Data jumlah rusak
    "P005, Kertas A4, 5" # Data tidak lengkap
]

def tes(data):
    baru = []
    for i in data:
        try:
            i_split = i.split(',')
            id_produk = i_split[0].strip()
            nama_produk = i_split[1].strip().title()
            jumlah = int(i_split[2].strip())
            harga = int(i_split[3].strip())
            baru.append({
                "id_produk": id_produk,
                "nama_produk": nama_produk,
                "jumlah": jumlah,
                "harga": harga
            })
        except ValueError:
            pass
    return baru

coba = tes(data_penjualan_h6)
print(json.dumps(coba, indent=4))
````
