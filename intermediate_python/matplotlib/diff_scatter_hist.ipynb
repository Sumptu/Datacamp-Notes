```python
import matplotlib.pyplot as plt
import numpy as np


rng = np.random.default_rng(seed=1)
# Misalkan ini adalah nilai ujian mahasiswa
grades = rng.normal(70, 10, 100)  # Data nilai dengan distribusi normal, rata-rata 70, standar deviasi 10, dan 100 data points

# Membuat histogram
plt.hist(grades, bins=10, edgecolor='black', alpha=0.7)

# Memberikan label
plt.title('Distribusi Nilai Ujian')
plt.xlabel('Nilai')
plt.ylabel('Frekuensi')

# Menampilkan histogram
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

rng = np.random.default_rng(123)

# Misalkan ini adalah data panjang jawaban (dalam kata) dan nilai ujian (dalam skala 0-100)
jawaban_panjang = rng.normal(50, 200, 100)  # panjang jawaban antara 50 dan 200 kata
nilai_ujian = rng.normal(70, 15, 100)  # nilai ujian acak dengan distribusi normal, rata-rata 70, standar deviasi 15

# Membuat scatter plot
plt.scatter(jawaban_panjang, nilai_ujian, alpha=0.7, color='blue')

# Memberikan label
plt.title('Panjang Jawaban vs Nilai Ujian')
plt.xlabel('Panjang Jawaban (Kata)')
plt.ylabel('Nilai Ujian')

# Menampilkan plot
plt.show()
```
