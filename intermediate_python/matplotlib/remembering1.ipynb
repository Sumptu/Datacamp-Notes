**Konteks:** Histogram standar (biasanya 10 bins) sudah memberikan gambaran, tetapi terkadang mengubah "resolusi" atau "granularitas" dari pengelompokan data bisa memberikan wawasan baru. Anda ingin melihat distribusi nilai dalam kelompok yang lebih besar.

4.  Tampilkan plotnya.
5.  Sebagai latihan interpretasi, tambahkan komentar di bawah kode Anda untuk menjawab pertanyaan: `"Dengan 5 bins, kelompok nilai manakah yang memiliki frekuensi tertinggi?"`

````python
plt.hist(nilai_ujian_4, bins=5)
plt.title("Distribusi Nilai Ujian Akhir")
plt.xlabel("Nilai Ujian")
plt.ylabel("Frekuensi (Jumlah Mahasiswa)")
plt.show()

# Kelompok nilai frekuensi tertinggi dari 85 sampai 90
````

# Tambahan Soal 5

Ketika kita mengatur bins=5, Matplotlib melihat nilai minimum (65) dan maksimum (100) dari data kita. Kemudian, ia membagi rentang tersebut (100 - 65 = 35) menjadi 5 kelompok dengan lebar yang sama (35 / 5 = 7).

Jadi, batas-batas (edges) dari setiap bin yang dibuat secara otomatis oleh Matplotlib adalah sekitar:

Bin 1: 65 - 72

Bin 2: 72 - 79

Bin 3: 79 - 86

Bin 4: 86 - 93

Bin 5: 93 - 100

Inilah mengapa batang histogramnya tidak sejajar persis dengan angka-angka cantik di sumbu x (70, 75, 80, dst.). Batang tertinggi di plot kita sebenarnya mewakili rentang nilai 86 hingga 93.

# Cara Membuatnya Lebih Jelas
Menentukan Batas Bin Secara Manual, Daripada memberikan angka (5), kita bisa memberikan sebuah list yang berisi batas-batas nilai yang kita inginkan. Ini memberi kita kendali penuh.



````python
# Menentukan batas bin yang jelas dan mudah dibaca
batas_bins = [60, 70, 80, 90, 100]

plt.hist(nilai_ujian_4, bins=batas_bins, edgecolor="black")
plt.title("Distribusi Nilai Ujian Akhir (Bins Didefinisikan)")
plt.xlabel("Nilai Ujian")
plt.ylabel("Frekuensi (Jumlah Mahasiswa)")
plt.show()
````

### **Soal 7: Menggabungkan Line Plot dan Scatter Plot**

2.  **Buat Line Plot (Tren Umum):**

### **Soal 8: Bekerja dengan Subplots**

**Konteks:** Anda ingin membandingkan dua metrik yang berbeda dari sebuah website: jumlah pengunjung harian dan durasi rata-rata sesi pengguna. Karena skala kedua metrik ini sangat berbeda, menampilkannya dalam satu plot akan sulit dibaca. Solusinya adalah menampilkannya dalam dua plot terpisah, tetapi dalam satu gambar (figure) yang sama, satu di atas yang lain.

**Tugas:**

1.  **Siapkan Figure:**
    * Buat sebuah *figure* baru dengan ukuran `(10, 8)` menggunakan `plt.figure(figsize=(10, 8))`.

2.  **Data:**
    ```python
    hari_8 = list(range(1, 11))
    pengunjung_harian_8 = [1200, 1350, 1400, 1250, 1500, 1600, 1800, 1750, 1900, 2100]
    durasi_sesi_menit_8 = [5.2, 5.5, 5.8, 5.4, 6.0, 6.2, 6.5, 6.8, 7.0, 7.2]
    ```

3.  **Buat Subplot Pertama (atas):**
    * Gunakan `plt.subplot(2, 1, 1)` untuk memilih area plot bagian atas. Angka `2, 1, 1` berarti: 2 baris, 1 kolom, plot ke-1.
    * Buat **line plot** untuk `pengunjung_harian_8`.
    * Beri judul `"Jumlah Pengunjung Harian"`.
    * Beri label sumbu-y `"Jumlah Pengunjung"`.

4.  **Buat Subplot Kedua (bawah):**
    * Gunakan `plt.subplot(2, 1, 2)` untuk memilih area plot bagian bawah.
    * Buat **line plot** untuk `durasi_sesi_menit_8` dengan warna `green`.
    * Beri judul `"Durasi Sesi Rata-Rata"`.
    * Beri label sumbu-x `"Hari ke-"` dan sumbu-y `"Menit"`.

5.  **Tampilkan Plot:**
    * Gunakan `plt.tight_layout()` agar judul dan label tidak tumpang tindih.
    * Tampilkan keseluruhan *figure* dengan `plt.show()`.

````python
hari_8 = list(range(1, 11))
pengunjung_harian_8 = [1200, 1350, 1400, 1250, 1500, 1600, 1800, 1750, 1900, 2100]
durasi_sesi_menit_8 = [5.2, 5.5, 5.8, 5.4, 6.0, 6.2, 6.5, 6.8, 7.0, 7.2]

plt.figure(figsize=(10, 8))
plt.subplot(2, 1, 1)
plt.plot(hari_8, pengunjung_harian_8)
plt.title("Jumlah Pengunjung Harian")
plt.ylabel("Jumlah Pengunjung")

plt.subplot(2, 1, 2)
plt.plot(hari_8, durasi_sesi_menit_8, color="green")
plt.title("Durasi Sesi Rata-Rata")
plt.xlabel("Hari ke-")
plt.ylabel("Menit")

plt.tight_layout()
plt.show()
````

### **Soal 9: Menyimpan Plot dan Menambah Anotasi Teks**

**Konteks:** Anda telah membuat visualisasi penting mengenai pertumbuhan pengguna. Anda perlu menyimpan plot ini sebagai file gambar (misalnya, PNG) untuk dimasukkan ke dalam laporan. Anda juga ingin menambahkan catatan teks langsung di plot untuk menyoroti titik tertinggi.

### **Soal 10: Memilih Plot yang Tepat dan Kustomisasi Lanjutan (Pie Chart)**

**Konteks:** Anda adalah seorang manajer produk untuk sebuah platform streaming. Anda ingin memvisualisasikan komposisi atau pangsa pasar dari berbagai genre konten yang ditonton oleh pengguna dalam satu bulan. Data seperti ini paling baik disajikan dengan **pie chart**.

**Tugas:**

1.  **Gunakan Data Berikut:**
    ```python
    genre_10 = ['Aksi', 'Komedi', 'Drama', 'Horor', 'Sains Fiksi']
    jumlah_penonton_10 = [4500, 3200, 2500, 1800, 1500]
    ```

2.  **Buat Pie Chart:**
    * Buat sebuah **pie chart** dari data di atas. Gunakan `jumlah_penonton_10` sebagai data utama dan `genre_10` sebagai labelnya.

3.  **Lakukan Kustomisasi Lanjutan:**
    * **`autopct`**: Tampilkan persentase setiap potongan (slice) hingga satu angka desimal di dalam pie chart. Gunakan `autopct='%1.1f%%'`.
    * **`explode`**: "Keluarkan" potongan dengan pangsa pasar terbesar (yaitu genre 'Aksi') sedikit dari pusat untuk menyorotnya. Buat sebuah tuple atau list `explode_setting = (0.1, 0, 0, 0, 0)`.
    * **`shadow`**: Tambahkan sedikit efek bayangan dengan `shadow=True`.
    * **`startangle`**: Putar pie chart sebesar 90 derajat agar potongan pertama dimulai dari atas, menggunakan `startangle=90`.

4.  **Lengkapi Plot:**
    * Tambahkan judul yang sesuai, misalnya "Komposisi Penonton Berdasarkan Genre".
    * Gunakan `plt.axis('equal')` untuk memastikan pie chart Anda berbentuk lingkaran sempurna.
    * Tampilkan plotnya.

````python
genre_10 = ['Aksi', 'Komedi', 'Drama', 'Horor', 'Sains Fiksi']
jumlah_penonton_10 = [4500, 3200, 2500, 1800, 1500]

plt.pie(
    jumlah_penonton_10,
    labels=genre_10,
    autopct="%1.1f%%",
    explode=(0.1, 0, 0, 0, 0),
    shadow=True,
    startangle=90,
)
plt.title("Komposisi Penonton Berdasarkan Genre")
plt.axis("equal")
plt.show()
````
